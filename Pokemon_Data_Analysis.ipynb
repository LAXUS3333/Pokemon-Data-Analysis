```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt



# Step 2: Calculate the average stats
# Get the averages for numerical stats
fire1_avg = fire1[['HP', 'Attack', 'Defense', 'SpAtk', 'SpDef', 'Speed']].mean()

# Step 3: Prepare the data for plotting
# Melt the original data to long format
fire1_melted = fire1.melt(
    id_vars=['Name'],
    value_vars=['HP', 'Attack', 'Defense', 'SpAtk', 'SpDef', 'Speed'],
    var_name='Stat',
    value_name='Value'
)

# Add a row for the averages
avg_rows = pd.DataFrame({
    'Name': ['Average'] * len(fire1_avg),
    'Stat': fire1_avg.index,
    'Value': fire1_avg.values
})

# Combine the original and average data
fire1_comparison = pd.concat([fire1_melted, avg_rows], ignore_index=True)

# Step 4: Plot the comparison
plt.figure(figsize=(14, 8))
sns.barplot(
    data=fire1_comparison,
    x='Name',
    y='Value',
    hue='Stat',
    palette='Set2'
)

# Customize the plot
plt.xticks(rotation=45, ha='right')
plt.title('Comparison of Fire-Type PokÃ©mon Stats with Averages (Generation 1)')
plt.ylabel('Stat Value')
plt.xlabel('PokÃ©mon Name')
plt.legend(title='Stat')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix



# Step 2: Select a specific PokÃ©mon and calculate averages
specific_pokemon_name = input("Write the name of fire type pokemon from Gen 1: ")
specific_pokemon = fire1[fire1['Name'] == specific_pokemon_name].iloc[0]
fire1_avg = fire1[['HP', 'Attack', 'Defense', 'SpAtk', 'SpDef', 'Speed']].mean()

# Step 3: Create labels for "Above Average" or "Below Average"
comparison_data = pd.DataFrame({
    'Stat': ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed'],
    'Specific_Pokemon': specific_pokemon[['HP', 'Attack', 'Defense', 'SpAtk', 'SpDef', 'Speed']].values,
    'Average': fire1_avg.values
})

# Assign categories
comparison_data['Specific_Pokemon_Label'] = comparison_data['Specific_Pokemon'] > comparison_data['Average']
comparison_data['Average_Label'] = comparison_data['Average'] > comparison_data['Average']  # Always False

# Map to categories
comparison_data['Specific_Pokemon_Label'] = comparison_data['Specific_Pokemon_Label'].map({True: "Above Average", False: "Below Average"})
comparison_data['Average_Label'] = comparison_data['Average_Label'].map({True: "Above Average", False: "Below Average"})

# Step 4: Build the confusion matrix
conf_matrix = confusion_matrix(
    comparison_data['Average_Label'],
    comparison_data['Specific_Pokemon_Label'],
    labels=["Above Average", "Below Average"]
)

# Step 5: Visualize the confusion matrix with the PokÃ©mon's name on the axis
plt.figure(figsize=(8, 6))
sns.heatmap(
    conf_matrix,
    annot=True,
    fmt="d",
    cmap="Blues",
    xticklabels=["Above Average", "Below Average"],
    yticklabels=["Above Average", "Below Average"]
)

plt.title(f"Confusion Matrix: {specific_pokemon_name} vs Average Stats")
plt.xlabel(f"{specific_pokemon_name}'s Stats")
plt.ylabel("Average Stats")
plt.show()
```

```python
specific_pokemon_name = input("Write the name of fire type pokemon from Gen 1: ")
specific_pokemon = fire1[fire1['Name'] == specific_pokemon_name].iloc[0]
fire1_avg = fire1[['HP', 'Attack', 'Defense', 'SpAtk', 'SpDef', 'Speed']].mean()

# Create the confusion matrix values (stat/average ratio)
comparison_matrix = pd.DataFrame({
    'Stat': ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed'],
    specific_pokemon_name: specific_pokemon[['HP', 'Attack', 'Defense', 'SpAtk', 'SpDef', 'Speed']].values,
    'Average': fire1_avg.values
})

# Calculate the ratio of stats to average
comparison_matrix['Ratio'] = comparison_matrix[specific_pokemon_name] / comparison_matrix['Average']

# Convert to matrix format for plotting
conf_matrix = comparison_matrix.pivot_table(
    values='Ratio',
    index='Stat',
    columns=specific_pokemon_name
)

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(
    conf_matrix.astype(float),  # Ensure numerical data
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    linewidths=0.5
)

# Customize the plot
plt.title(f"Stat Comparison: {specific_pokemon_name} vs Average Stats")
plt.xlabel(f"{specific_pokemon_name}'s Stats")
plt.ylabel("Stat Categories")
plt.tight_layout()
plt.show()
```
